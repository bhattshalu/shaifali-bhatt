```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt
def relu(x):
  return (np.maximum(x,0))

x = np.linspace(-10,10,200)
y = np.linspace(-10,10,200)



f1 = x*x*x + x*x - x - 1

plt.plot(x,f1)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),y, color = 'black')
plt.grid()
```

```python
x = np.arange(-2,2,.2)


n1 = relu(-5*x -7.7)
n2 = relu(-1.2*x -1.3)
n3 = relu(1.2*x + 1)
n4 = relu(1.2*x -.2)
n5 = relu(2*x -1.1)
n6 = relu(5*x - 5)

z = -n1 -n2 -n3 +n4 +n5 +n6

plt.plot(x,z)
plt.plot(x,np.zeros(x.shape), color = 'black')
plt.plot(np.zeros(x.shape),x, color = 'black')
plt.show()
```
